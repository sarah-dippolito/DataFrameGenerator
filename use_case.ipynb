```python
from dataframe_generator import DataFrameGenerator

# use case 1 (orders)

gen_df = DataFrameGenerator(nrows=10, random_state=12)

# add cols
gen_df.add_column("order_id", ("id", 0))
gen_df.add_column("customer_code", ("prefix_id", "CUST", 4, False))
gen_df.add_column("customer_id", ("alphanumeric_id", 8, False))
gen_df.add_column("customer_initials", ("string", 3, "upper"))
gen_df.add_column("num_products", ("int", 1, 20))
gen_df.add_column("price", ("correlated", "num_products", 3, 0.5))
gen_df.add_column("shipping", ("int_prob", [1,1.5,2], [0.25,0.5,0.25]))
gen_df.add_column("delivered", ("bernoulli", 0.65))
gen_df.add_column("pct_paid", ("percent", 10,2))
gen_df.add_column("num_payment_methods", ("int_prob", [1,2,3], [0.9, 0.06, 0.04]))
gen_df.add_column("order_date", ("datetime", '2025-01-01', '2025-12-31'))
gen_df.add_column("delivered_from", ("cat_prob", ['Amazon', 'Website', 'Store'], [0.4, 0.5, 0.1]))
gen_df.add_column("survey_completion", ("beta", 3, 5, 100, {'missing':0.7}))  #missing used here
df = gen_df.build()


df.head(10)
```

```python
from dataframe_generator import DataFrameGenerator

# use case 2 (college)

gen_df = DataFrameGenerator(nrows=10, random_state=1)

# add cols
gen_df.add_column("student_id", ("prefix_id", "STU", 4))
gen_df.add_column("fist_initial", ("string", 1, "upper"))
gen_df.add_column("last_initial", ("string", 1, "upper"))
gen_df.add_column("major", ("cat", ['Psychology', 'Finance', 'Biology', 'Fashion', 'Art', 'Political Science', 'Math', 'Marketing', 'Engineering', 'Computer Science']))
gen_df.add_column("CUM_GPA", ("float", 0,4))
gen_df.add_column("major_GPA", ("normal_clip", 3.1, 0.3, 0, 4))
gen_df.add_column("graduated", ("bool", ))
gen_df.add_column("birth_date", ("date", '2002-01-01', '2006-01-01'))
df = gen_df.build()

df['CUM_GPA'] = round(df['CUM_GPA'], 2)
df['major_GPA'] = round(df['major_GPA'], 2)
df.head(10)
```
